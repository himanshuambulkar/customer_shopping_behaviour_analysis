```python
# loading dataset using pandas
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
# summary statistics using describe.
df.describe(include="all") 
```

```python
# check the number of null values in each columns.
df.isnull().sum() 
```

```python
# filled null values in the rows using median.
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))

```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower() # here we change the column names to lower case
df.columns=df.columns.str.replace(' ','_') # here we replace spaces in columns names with underscores.
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#create a new column age_group
labels=["young adult","adult","middle-aged","senior"]
df["age_group"] = pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(30)
```

```python
#create a new column frequency_of_purchases_days
frequency_mapping={'Fortnightly':14,'Weekly':7,'Annually':365,'Quarterly':90,'Bi-Weekly':14,
                   'Monthly':30,'Every 3 Months':90}
# here we convert the text column into numeric using map function.
df['frequency_of_purchases_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','frequency_of_purchases_days']].head(30)
```

```python
df[['promo_code_used','discount_applied']].head(10)
```

```python
(df['promo_code_used'] == df['discount_applied']).all()
```

```python
# drop promo code 
df=df.drop('promo_code_used',axis=1)
```
